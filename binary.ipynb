```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
cc = pd.read_csv('cleaned_credit_card.csv')
app = pd.read_csv('app_record_ml.csv')
```

```python
cc.columns
```

```python
app.columns
```

```python
df = pd.merge(cc, app, on='ID')
del df['Unnamed: 0']
```

```python
df
```

```python
del df['ID']
df.columns
```

```python
df = df.dropna()
```

```python
df
```

```python
X = df.drop(['0', '1', '2', '3', '4', '5', 'X', 'C'], axis=1)
y = df[['0', '1', '2', '3', '4', '5', 'X', 'C']]
```

```python
X
```

```python
y
```

```python
#y['score'] = y['0']+ (1* y['1']) + (1*y['2'])+ (1*y['3']) + (1*y['4']) + (1*y['5'])
```

```python
y['score'] = y['2']#/y['3']# + y['C']
```

```python
y = y.fillna(0)
```

```python
y = y['score']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_classifier = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr')

# Train the model
log_classifier.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = log_classifier.predict(X_test)
print(f" Logistic regression model accuracy: {accuracy_score(y_test,y_pred):.3f}")
```

```python
print(f"Training Data Score: {log_classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {log_classifier.score(X_test, y_test)}")
```

```python
from sklearn.tree import DecisionTreeClassifier

#clf = clf.fit(X_train, y_train)
#clf.score(iris.data, iris.target)

tree = DecisionTreeClassifier(max_depth=50,
                               min_samples_split=2,
                               random_state=10)
tree.fit(X_train, y_train)
print(f"Training Data Score: {tree.score(X_train, y_train)}")
print(f"Testing Data Score: {tree.score(X_test, y_test)}")
```

```python
tree_scaled = DecisionTreeClassifier(max_depth=200,
                               min_samples_split=2,
                               random_state=10)
tree_scaled.fit(X_train_scaled, y_train)
print(f"Training Data Score: {tree_scaled.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {tree_scaled.score(X_test_scaled, y_test)}")
```
